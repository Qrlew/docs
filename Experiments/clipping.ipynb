## New methodology

We will assume the count for a user will be concentrated as a $\beta(1,\alpha)$ on one vector and that the vector is concentrated on one dimension (group).

The qth quantile of the count is : $C_q = N\cdot\left(1-(1-q)^\alpha\right)$

When $\alpha$ is small, we have $C_q = -\alpha\cdot N\cdot\ln(1-q)$

```python
import numpy as np
import matplotlib.pyplot as plt

def quantile_plot(size=10000, alpha=1):
    n = 1000
    q = np.linspace(0,1,n)
    return plt.plot(q, np.ceil(size*(1-(1-q)**alpha)))

for a in [0.01, 0.1, 0.5, 1]:
    quantile_plot(alpha=a)
plt.show()

def concentration_plot(size=10000, q=0.5):
    n = 1000
    alpha = np.linspace(0,1,n)
    return plt.plot(alpha, np.ceil(size*(1-(1-q)**alpha)))

for q in [0.5, 0.75, 0.9, 0.95]:
    concentration_plot(q=q)
plt.show()
```

```python
plt.xlim(0,0.9)
plt.ylim(0,10)
for s in [100, 1000, 10000, 100000]:
    quantile_plot(alpha=1/s)
plt.show()
```
